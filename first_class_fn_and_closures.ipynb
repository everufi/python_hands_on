```python
def func(x):
  return x+1
def apply_twice(func,x):
  return func(func(x))

print(apply_twice(func,5))
```

```python
def make_multiplier(k):
  def multiplier(x):
    return x*k
  return multiplier

s1=make_multiplier(2)
print(s1(7))
```

```python
def counter():
  count=0
  def next():
    nonlocal count
    count+=1
    return count
  return next
c1=counter()
print(c1())
print(c1())
```

```python
def make_accumulator(start=0):
  total=start

  def accumulator(x):
    nonlocal total
    total+=x
    return total
  return accumulator
a1=make_accumulator(5)
print(a1(7))
print(a1(-3))
```
