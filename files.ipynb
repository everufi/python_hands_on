```python
books = [
    {"title": "Book1", "author": "Author1", "price": 300},
    {"title": "Book2", "author": "Author2", "price": 450}
]

import csv
with open("bookstore.csv","w") as file:
  dict1=csv.DictWriter(file,fieldnames=["title","author","price"])
  dict1.writeheader()
  dict1.writerows(books)
with open("bookstore.csv","r") as file:
  dict1=csv.DictReader(file)
  for row in dict1:
    print(row)
```

```python
books = [
    {"title": "Book1", "author": "Author1", "price": 100},
    {"title": "Book2", "author": "Author2", "price": 50},
    {"title": "Book3", "author": "Author3", "price": 30}
]

import csv
with open("bookstore.csv","w") as file:
  dict1=csv.DictWriter(file,fieldnames=["title","author","price"])
  dict1.writeheader()
  dict1.writerows(books)

with open("bookstore.csv","r") as file:
  dict1=csv.DictReader(file)

  for row in dict1:
    if(int(row["price"]) >= 50):
      print(row)
```
