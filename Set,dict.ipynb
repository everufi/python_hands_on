```python
d1={}
d2={}
d3={}
n1=int(input())
n2=int(input())
for i in range(n1):
  key,value=input().split()
  d1[key]=value
for i in range(n2):
  key,value=input().split()
  d2[key]=value
for key in d1:
  if key in d2:
    d3[key]=int(d1[key])+int(d2[key])
for key in d3:
  print(key,d3[key])
```

```python
l1=list(map(int,input().split()))
dup=set()
l2=[]
for i in l1:
  if i not in dup:
    l2.append((i,l1.count(i)))
    dup.add(i)
print(l2)
```
