```python
from abc import ABC,abstractmethod

class PaymentGateway(ABC):

  @abstractmethod
  def pay(self,amount):
    pass

  def refund(self,amount):
    pass

class UPI(PaymentGateway):
  def pay(self,amount):
    print(f"Paid {amount} using UPI")

  def refund(self,amount):
    print(f"Refunded {amount} using UPI")

class CreditCard(PaymentGateway):
  def pay(self,amount):
    print(f"Paid {amount} using Credit Card")

  def refund(self,amount):
    print(f"Refunded {amount} using Credit Card")

c=CreditCard()
c.pay(1000)
```

```python
from abc import ABC,abstractmethod
class Employee(ABC):
  def __init__(self,name,emp_id):
    self.name=name
    self.emp_id=emp_id
  @abstractmethod
  def calculate_salary(self):
    pass

class Full_time_employee(Employee):
  def __init__(self,name,emp_id,salary):
   super().__init__(name,emp_id)
   self.salary=salary

  def calculate_salary(self):
    return self.salary

class Intern(Employee):
  def __init__(self,name,emp_id,stipend):
    super().__init__(name,emp_id)
    self.stipend=stipend

  def calculate_salary(self):
    return self.stipend

i=Intern("Eva", "01", 10000)
e=Full_time_employee("Rufi", "01", 100000)
print(f"Intern Salary: {i.calculate_salary()}")
print(f"Full-time Employee Salary: {e.calculate_salary()}")
```

```python
from abc import ABC,abstractmethod
class Filehandler(ABC):
  @abstractmethod
  def read(self):
    pass

  @abstractmethod
  def write(self):
    pass

class Textfile(Filehandler):
  def read(self):
    print("Reading text file")

  def write(self):
    print("Writing text file")

class CSVfile(Filehandler):
  def read(self):
    print("Reading CSV file")

  def write(self):
    print("Writing CSV file")

t=CSVfile()
t.read()
t.write()
```

```python
from abc import ABC,abstractmethod
class Validator(ABC):
  @abstractmethod
  def validate(self):
    pass

class Emailvalidator(Validator):
  def validate(self,data):
    if " " in data:
      return False

    if data.count("@")!=1:
      return False

    name,mail=data.split("@")

    if name==" " or mail==" ":
      return False

    if "." not in mail:
      return False


    return True

class PWvalidator(Validator):
  def validate(self,data):
    if len(data)<8:
      return False

    if not any(char.isdigit() for char in data):
      return False

    if not any(char.isupper() for char in data):
      return False

    if not any(char.islower() for char in data):
      return False


    return True

e=Emailvalidator()
print(e.validate("evie_rufi_rilcy@gmail.com"))
```
