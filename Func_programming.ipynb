```python
marks=[45,78,88,32,67,49,90]
from functools import reduce
l1=list(map(lambda x:x*1.05,marks))
l2=list(filter(lambda x:x>60,l1))
l3=reduce(lambda x,y:x+y,l2,0)
avg=l3/len(l2)
print(l1)
print(l2)
print(l3)
print(avg)

```

```python
prices=list(map(int,input().split()))
from functools import reduce
disc=list(map(lambda x:x-(x*0.1),prices))
fil=list(filter(lambda x:x>=200,disc))
total=reduce(lambda x,y:x+y,fil,0)
print(total)
```

```python
from functools import reduce
scores = list(map(float, input().split()))
weights = list(map(float, input().split()))
weightsum = reduce(lambda acc, pair: acc + pair[0] * pair[1],zip(scores, weights),0)
total=sum(weights)
score = weightsum / total if total != 0 else 0
print(score)
```
