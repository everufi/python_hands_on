```python
l1=list(map(int,input().split()))
duplicate=set()
result=[]
for i in reversed(l1):
  if i not in duplicate:
    duplicate.add(i)
    result.append(i)
result.reverse()
print(result)
```

```python
l1=list(map(str,input().split()))
result=[]
for i in range(0,len(l1)-1):
  result.append((l1[i],l1[i+1]))
print(result)
```

```python
l1=list(map(int,input().split()))
distinct=list(set(l1))
distinct.sort()
print(distinct[-2])
```

```python
lst=list(map(int,input().split()))
n=len(lst)
k=int(input())
k=k%n
for i in range(k):
  num=lst.pop()
  lst.insert(0,num)
print(lst)
```

```python
lst1=list(map(int,input().split()))
lst2=list(map(int,input().split()))
result=[]
for i in range(min(len(lst1),len(lst2))):
  result.append(lst1[i])
  result.append(lst2[i])
print(result)
```

```python
correct_pw='rec#123'
pw=input()
attempt=3

while attempt>0:
  if pw==correct_pw:
    print("Login done")
    break
  else:
    attempt-=1
    print("Incorrect password")
    pw=input()
if attempt<=0:
  print("no attempt left")
```
