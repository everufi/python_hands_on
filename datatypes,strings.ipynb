```python
amount=int(input())
final=0
if amount<0:
  print('Invalid')
elif amount>0 and amount<1000:
  print("No discount available")
elif amount>=1000 and amount<=4999:
  final=amount-((5/100)*amount)
else:
  final=amount-((10/100)*amount)
print(final)
```

```python
text=input()
words=text.split()
abb=['AI','ML','ISRO','NASA']
result=[]
for i in words:
  if i in abb:
    result.append(i)
  else:
    result.append(i.capitalize())

print(" ".join(result))

```

```python
correct_pw='rec#123'
pw=input()
attempt=3

while attempt>0:
  if pw==correct_pw:
    print("Login done")
    break
  else:
    attempt-=1
    print("Incorrect password")
    pw=input()
if attempt<=0:
  print("no attempt left")
```
