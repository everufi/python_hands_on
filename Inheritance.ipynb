```python
class Writer:
    def write(self):
        return "Writing content..."


class Reader:
    def read(self):
        return "Reading content..."


class Editor(Writer, Reader):
    def edit(self):
        return "Editing content..."
e1=Editor()
print(e1.read())
print(e1.write())
print(e1.edit())
```

```python
class Employee:

    def __init__(self, salary, bonus):
        self._salary = salary      # Protected
        self.__bonus = bonus       # Private

    def get_bonus(self):
        return self.__bonus

class Manager(Employee):

    def display_salary(self):
        # Can access protected attribute
        print("Salary:", self._salary)

    def display_bonus(self):
        # Trying to access private attribute directly
        try:
            print("Bonus:", self.__bonus)
        except AttributeError:
            print("Cannot access private attribute __bonus directly")
m = Manager(50000, 10000)

m.display_salary()
m.display_bonus()


print("Bonus via getter:", m.get_bonus())
```

```python
class Logger:

    def log(self, message):
        print(f"[LOG]: {message}")


class FileLogger(Logger):

    def log(self, message):
        # Call parent log method
        super().log(message)
        print(f"Writing '{message}' to file...")

l1 = FileLogger()
l1.log("System started")
```

```python
class Vehicle:

    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} vehicle is starting...")

    def stop(self):
        print(f"{self.brand} vehicle is stopping...")


class Car(Vehicle):

    def __init__(self, brand, model):
        super().__init__(brand)   # Call Vehicle constructor
        self.model = model

    def honk(self):
        print(f"{self.brand} {self.model} says Beep Beep!")


class ElectricCar(Car):

    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)   # Call Car constructor
        self.battery_capacity = battery_capacity

    def charge(self):
        print(f"{self.brand} {self.model} is charging...")
tesla = ElectricCar("Tesla", "Model 3", "75kWh")

# Method from Vehicle
tesla.start()

# Method from Car
tesla.honk()

# Method from ElectricCar
tesla.charge()

tesla.stop()
print(ElectricCar.__mro__)
```
